## Data Import and first Inspection

1. __Import__ the movies dataset from the CSV file "movies_complete.csv". __Inspect__ the data.

__Some additional information on Features/Columns__:

* **id:** The ID of the movie (clear/unique identifier).
* **title:** The Official Title of the movie.
* **tagline:** The tagline of the movie.
* **release_date:** Theatrical Release Date of the movie.
* **genres:** Genres associated with the movie.
* **belongs_to_collection:** Gives information on the movie series/franchise the particular film belongs to.
* **original_language:** The language in which the movie was originally shot in.
* **budget_musd:** The budget of the movie in million dollars.
* **revenue_musd:** The total revenue of the movie in million dollars.
* **production_companies:** Production companies involved with the making of the movie.
* **production_countries:** Countries where the movie was shot/produced in.
* **vote_count:** The number of votes by users, as counted by TMDB.
* **vote_average:** The average rating of the movie.
* **popularity:** The Popularity Score assigned by TMDB.
* **runtime:** The runtime of the movie in minutes.
* **overview:** A brief blurb of the movie.
* **spoken_languages:** Spoken languages in the film.
* **poster_path:** The URL of the poster image.
* **cast:** (Main) Actors appearing in the movie.
* **cast_size:** number of Actors appearing in the movie.
* **director:** Director of the movie.
* **crew_size:** Size of the film crew (incl. director, excl. actors).

## The best and the worst movies...

2. __Filter__ the Dataset and __find the best/worst n Movies__ with the

- Highest Revenue
- Highest Budget
- Highest Profit (=Revenue - Budget)
- Lowest Profit (=Revenue - Budget)
- Highest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10) 
- Lowest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10)
- Highest number of Votes
- Highest Rating (only movies with 10 or more Ratings)
- Lowest Rating (only movies with 10 or more Ratings)
- Highest Popularity

```python
df_best = df[["poster_path", "title", "budget_musd", "revenue_musd",
              "vote_count", "vote_average", "runtime", "popularity"]].copy()
df_best
```

```python
df_best['profit_musd'] = df.revenue_musd.sub(df.budget_musd)
df_best['return'] = df.revenue_musd.div(df.budget_musd)
```

```python
df_best.columns = ['', 'Tittle', 'Budget', 'Revenue', 'votes', 'Average rating', 'RunTime','Popularity', 'Profit', 'ROI']
```

```python
df_best.set_index('Tittle', inplace=True)
```

```python
subset = df_best.iloc[:5, :3]
```

```python
df_best.sort_values(by='Average rating', ascending= False)
```

```python
df_best.sort_values(by='votes', ascending= False)
```

```python
df_best.Budget.fillna(0, inplace = True)
df_best.votes.fillna(0, inplace = True)
df_best.RunTime =df_best.RunTime.fillna(df_best.RunTime.mean())
```

```python
df_best.RunTime.mean()
```

```python
df_best.info()
```

```python
def best_worst (n, by, ascending = False, min_bud = 0, min_votes = 0, run_time = 0):
    df2 = df_best.loc[(df_best.Budget >= min_bud) & (df_best.votes >= min_votes)&(df_best.RunTime >= run_time),["", by]].sort_values(by = by, ascending = ascending).head(n).copy()
    return HTML(df2.to_html(escape=False))
```

3. __Filter__ the Dataset for movies that meet the following conditions:

__Search 1: Science Fiction Action Movie with Bruce Willis (sorted from high to low Rating)__

```python
df.columns = df.columns.str.strip()
Scifi_action_bruce = (df.genres.str.contains('Action') 
                      & df.genres.str.contains('Science Fiction') 
                      & df.cast.str.contains('Bruce Willis'))
sorted_df = df.loc[Scifi_action_bruce, ['title', 'poster_path','vote_average']].sort_values(by='vote_average', ascending=False).set_index('title')
HTML(sorted_df.to_html(escape = False))
```

__Search 2: Movies with Uma Thurman and directed by Quentin Tarantino (sorted from short to long runtime)__

```python
df.head()
```

```python
df.columns = df.columns.str.strip()
Scifi_action_uma = (  df.director.str.contains('Quentin Tarantino') 
                      & df.cast.str.contains('Uma Thurman'))
sorted_df_uma = df.loc[Scifi_action_uma, ['title', 'poster_path','runtime']].sort_values(by='runtime').set_index('title')
HTML(sorted_df_uma.to_html(escape = False))
```

__Search 3: Most Successful Pixar Studio Movies between 2010 and 2015 (sorted from high to low Revenue)__

```python
df.info()
```

```python
df.release_date = pd.to_datetime(df.release_date)
```

```python
df.columns = df.columns.str.strip()
scifi_action_pix = (df['production_companies'].str.contains('Pixar.*Studios', case=False, na=False) & 
                    (df['release_date'].dt.year >= 2010) & 
                    (df['release_date'].dt.year <= 2015))
sorted_df_pix = df.loc[Scifi_action_uma, ['title', 'poster_path','production_companies','revenue_musd']].sort_values(by='revenue_musd', ascending = False).set_index('title')
HTML(sorted_df_pix.to_html(escape = False))
```

__Search 4: Action or Thriller Movie with original language English and minimum Rating of 7.5 (most recent movies first)__

```python
df.columns = df.columns.str.strip()
scifi_action_AT = (
    (df['genres'].str.contains('Action', case=False, na=False)) |  
    (df['genres'].str.contains('Thriller', case=False, na=False))
) & (df['original_language'].str.contains('en', case=False, na=False)) & (df['vote_average'] >= 7.5) & (df['vote_count'] >= 10)
sorted_df_AT = df.loc[scifi_action_AT, ['title', 'poster_path','genres', 'vote_average', 'vote_count', 'release_date']] \
    .sort_values(by='vote_average', ascending=False) \
    .set_index('title')
HTML(sorted_df_AT.head(10).to_html(escape=False))
```

# Most sucessfull Actors

```python
df.cast
```

```python
df.set_index('id', inplace= True)
```

```python
df.cast
```

```python
act = df.cast.str.split("|", expand = True)
act
```

```python
act = act.stack().reset_index(level = 1, drop = True).to_frame()
act
```

```python
act.columns = ['Actors']
```

```python
act = act.merge(df[['title','revenue_musd','vote_average','popularity']],how = 'left',left_index=True, right_index=True)
```

```python
act
```

```python
act.Actors.nunique()
```

```python
act.Actors.unique()
```

```python
act.Actors.value_counts().head(20)
```

```python
plt.figure(figsize = (12, 8))
act.Actors.value_counts().head(20).plot(kind='bar', fontsize = 15)
plt.title("Most Active Actors",fontsize = 20)
plt.ylabel("Number of Movies", fontsize = 15)
plt.show()
```

```python
agg = act.groupby("Actors").agg(Total_Revenue = ("revenue_musd", "sum"), 
                               Mean_Revenue = ("revenue_musd", "mean"),
                               Mean_Rating = ("vote_average", "mean"), 
                               Mean_Pop = ("popularity", "mean"), 
                               Total_Movies = ("Actors", "count"))
```

```python
agg.nlargest(10,'Total_Revenue')
```

```python
agg.nlargest(20, 'Total_Movies')
```

```python
plt.figure(figsize = (12, 8))
agg.Total_Revenue.nlargest(10).plot(kind='bar', fontsize = 15)
plt.title("Total Revenue",fontsize = 20)
plt.ylabel("Revenue (in MUSD)", fontsize = 15)
plt.show()
```
